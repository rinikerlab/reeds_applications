```python
def plot_free_energies_forSI(exp, sim, labels_sim, title=None, colors=None, markers=None,
                       two_greyed_regions=True, relative=False, manual_lims=None):
    """
    This function will plot the results in a square scatter plot
    
    """

    active_qualitative_map_mligs = lambda num_ligs: plt.cm.viridis(np.linspace(0,1,num_ligs))
    active_qualitative_map_mligs = lambda num_ligs: plt.cm.cividis(np.linspace(0,1,num_ligs))
    color_blind_frienfly = active_qualitative_map_mligs(5)

    def plot_errorbar(ax, x, y, y_err, color, label, zorder, alpha, marker = 's'):
        markers, caps, bars = ax.errorbar(x, y, y_err, fmt=marker, markersize = 6,
                                          color = color, mec = color, 
                                          ecolor = color, label = label,
                                          capthick=1.5, capsize=4, linewidth = 3,
                                          zorder=zorder, alpha = 1
                                         )
        [bar.set_alpha(alpha) for bar in bars]
        [cap.set_alpha(alpha) for cap in caps]

    xmin = np.min(exp)
    xmax = np.max(exp)

    fig, ax = plt.subplots(ncols=1, nrows=1, figsize =[9,9])

    #if colors is None:
    colors = ['black', 'orange', 'skyblue', 'darkorange']

    if markers is None:
        markers = ['D'] * len(labels_sim)

    if len(sim) == 2: # We we have gaff and openFF
        plot_errorbar(ax, exp, sim[0][0], sim[0][1], color = colors[0],
                      label = labels_sim[0], zorder = 10, alpha = 1, marker = 's')

        plot_errorbar(ax, exp + 0.1 , sim[1][0] + 0.1, sim[1][1], color = colors[1],
              label = labels_sim[1], zorder = 10, alpha = 0.75, marker = 's')



    leg = ax.legend(loc = 'best', fontsize=20, fancybox=True)
    leg.get_frame().set_edgecolor('black')

    x = np.linspace(-100, 100, 50)

    ax.plot(x, x, color = 'black')
    ax.plot(x, x-4.185, color = 'lightgrey')
    ax.plot(x, x+4.185, color = 'lightgrey')
    #
    ax.fill_between(x, x-4.185, x+4.185, color='lightgrey')

    if two_greyed_regions:
        ax.fill_between(x, x-2*4.185, x+2*4.185, color='lightgrey', alpha=0.5)

    if title is not None:
        ax.set_title(title, fontsize = 18)

    ax.set_xlabel(r'$\Delta G_{bind}$ - Experimental [kJ/mol]', fontsize = 24)
    ax.set_ylabel(r'$\Delta G_{bind}$ - Simulation [kJ/mol]', fontsize = 24)

    # Automatically find the value in x closest to largest and smallest values
    ax.set_ylim([-46.5, -29.5])
    ax.set_xlim([-46.5, -29.5])
    
    # Formatting
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(1.5)

    ax.tick_params(axis="y",direction="in", length = 6, width = 1.5, labelsize=18)
    ax.tick_params(axis="x",direction="in", length = 6, width = 1.5, labelsize=18)
    
    ax.set_yticks(np.arange(-50, -25, 5))
    ax.set_xticks(np.arange(-50, -25, 5))
    
    return fig, ax
```

```python
labels_extra = [f'QligFep\nMUE = 2.02 kJ/mol', f'FEP+\nMUE = 1.72 kJ/mol']


# Here we can add / remove to plot what we want 

fig, ax = plot_free_energies_forSI(dgs_exp, [qlip_fep, dgs_fep_plus], labels_sim = labels_extra, 
                         title =None, two_greyed_regions=True, manual_lims = [-46.5, -29.5], colors = ['royalblue', 'firebrick'])



ax.set_ylim([-46.5, -29.5])
ax.set_xlim([-46.5, -29.5])
    
```
