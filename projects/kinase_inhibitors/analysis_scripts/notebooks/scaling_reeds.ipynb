```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
from matplotlib import cm
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize =[7,6])

%matplotlib inline
fig.show()

font = {'size'   : 14}

mpl.rc('font', **font)


n_seeds = 1 

n_ligands = np.arange(2, 16)

ti_linear = 21 * (n_ligands-1) * 5 * n_seeds 
ti_quadratic = 21 * (n_ligands * (n_ligands))/2 * 5 * n_seeds

ax.plot(n_ligands, ti_linear, c='darkgrey', lw = 3, zorder = 2)
ax.plot(n_ligands, ti_quadratic, c='darkgrey', lw = 3, zorder =2, label = 'Theoretical estimate\nFEP/TI')

ax.plot(n_ligands, (n_ligands)*np.log(n_ligands) * 21 * 5 * n_seeds , c='black', 
        ls = '--', lw = 3, label = r'Pitman et al. $\mathcal{O}(n\log{n})$ ', zorder =1)


ax.fill_between(n_ligands, ti_linear, ti_quadratic, color = 'lightgrey', zorder =0)



get_colors = lambda num_ligs: plt.cm.tab20b(np.linspace(0,1,num_ligs))

colors =  plt.cm.tab20b( (4./3*np.arange(20*3/4)).astype(int) )[1::3]

marker ='d'
t = mpl.markers.MarkerStyle(marker=marker)
t._transform = t.get_transform().rotate_deg(90)

ax.scatter([5, 13], [184+160*n_seeds, 333+160*n_seeds], label = 'CHK1',color = colors[0], marker = t, s = 200, zorder = 5)


ax.scatter(7, 266+160*n_seeds, label = 'PAK', color = colors[3], marker = t, s = 200, zorder = 5)


ax.scatter(13, 141+160*n_seeds, label = 'PIM', color = colors[2], marker = t, s = 200, zorder = 5)

ax.scatter(6, 300+160*n_seeds, label = 'NIk', color = colors[1], marker = t, s = 200, zorder = 5)


ax.legend(loc='upper left', edgecolor='white', fontsize = 16)

ax.set_xticks(n_ligands[::2])
ax.set_xticklabels(n_ligands[::2],fontsize=18)
ax.set_xlabel('Number of ligands', fontsize =18)

ax.set_ylabel(r'Cumulative simulation time [$\mu$s]', fontsize = 18)

ticks = ax.get_yticks()

ax.set_yticklabels(np.array(ticks/1000, dtype=int), fontsize = 18)

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
ax.tick_params(width=2)

```
