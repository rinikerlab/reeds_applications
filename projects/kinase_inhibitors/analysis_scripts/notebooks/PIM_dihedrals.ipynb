Idea behind this notebook is to extract the dihedral angles of the -R group to see the motion in the simulations, and plot those

```python
def get_props(core_atoms, rgroup_atoms):
    """
    Will write the names based on the atom numbers of th eatoms in the torsion to monitor
    """
    i, j = core_atoms
    props = []
    
    for k, l in rgroup_atoms:
        props.append('t%1:' + str(i) +','+ str(j)+','+ str(k) +','+str(l))
    
    return props

def extract_all_props(topo, trajs, props, out_name, pbc = "r cog", time = "0 2", ):
    """
    Extract all properties given as input in all trajectories
    (good for RE-EDS)
    """
    if os.path.exists(out_name):
        print ('file exists. exit')
        return
        
    if isinstance(trajs, list):
        trajs = " ".join(trajs)
    
    command = "tser @topo " + topo + " @traj " + trajs + " @pbc " + pbc + " @time " + time 
    command += " @prop " + " ".join(props) + " | grep -v '#' > " + out_name
        
    print (command)
    os.system(command)
    print ('\n\n')

def extract_sep_props(topo, trajs, props, out_name, pbc = "r cog", time = "0 2", ):
    """
    Extract the properties given in the argument props separately (for step A)
    
    """
    for i, (prop, traj) in enumerate(zip(props, trajs)):
        command = "tser @topo " + topo + " @traj " + traj + " @pbc " + pbc + " @time " + time 
        command += " @prop " + prop + " | grep -v '#' > " + out_name + '_' + str(i+1) + '.csv'
        
        print (command)
        os.system(command)
        print ('\n\n')
        
def find_trajs_from_chunks(in_path, s = 1):
    files = [f for f in glob.glob(in_path+'/*/*.trc.gz') if 'eq' not in f]
    
    # keep only the replicas for the correct s 
    files = [f for f in files if int(f.split('/')[-1].split('.')[0].split('_')[-1]) == s]
    
    # sort 
    return sorted(files, key = lambda x: int(x.split('/')[-2].split('_')[-1]))            
    
        
def extract_data_independent_eds(path_data, prefix): 

    path_csvs = sorted(glob.glob(path_data+ '/' + prefix + '*.csv'), key = lambda x : int(x.split('.')[0].split('_')[-1]))
    
    time = None
    data = []
    
    for i, csv in enumerate(path_csvs):
        vec = np.genfromtxt(csv).T
        if i == 0:
            time = vec[0] / 1000 # to ns
        data.append(vec[1:])
        
    return time, data

def extract_data(path_csv): 
    vec = np.genfromtxt(path_csv).T    
    # time, dihedrals
    return vec[0]/1000, mod2pi(vec[1:])


def extract_data_other(path_csv): 

    time = None
    data = []    

    vec = np.genfromtxt(path_csv).T
    if i == 0:
        time = vec[0] / 1000 # to ns
    data.append(vec[1:])
        
    return time, data


def mod2pi(v):
     return np.degrees(np.mod(np.radians(v), 2*np.pi)) 
    
def plot_dof_timeseries(time, dihedrals, outfile = None, title = None, idx_min = None):
    """
    Update this with the minimum energy point highlighted!!!!
    
    """
    nstates = len(dihedrals)
    
    active_qualitative_map_mligs = lambda num_ligs: plt.cm.viridis(np.linspace(0,1,num_ligs))
    
    colors = active_qualitative_map_mligs(nstates)
    # Split the different states in subplots:

    ncols = 4 if nstates > 11 else 3
    nrows = int(np.ceil(nstates/ncols))

    figsize = [3*ncols, 3*nrows]
    fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=figsize)
    
    if title is None:
        fig.suptitle('Dihedrals for optimized states', fontsize = 16)
    else:
        fig.suptitle(title, fontsize = 16)
    
    
    row_num = 0 # this starts at 0, and gets updated when col_num = ncols
    col_num = 0

    for i, dihedral in enumerate(dihedrals):
        if col_num == ncols-1: row_num +=1
        col_num = i % ncols
        state_num = i+1
        
        axes[row_num, col_num].scatter(time, mod2pi(dihedral), s = 2, color = colors[i])
        axes[row_num, col_num].set_ylim([0, 360])
        axes[row_num, col_num].set_title('state ' + str(i+1))
        
        # Set the ticks to make sense:
        
        axes[row_num, col_num].set_yticks([0, 60, 120, 180, 240, 300, 360])
                
        if idx_min is not None:
              axes[row_num, col_num].scatter(time[idx_min[i]-1], mod2pi(dihedral)[0][idx_min[i]-1], 
                                             color = 'black', marker ='*', s = 32)
        
    
    # Add labels
    for i in range(ncols):
        axes[nrows-1, i].set_xlabel('time [ns]')
    
    for i in range(nrows):
        axes[i, 0].set_ylabel('dihedral angle [deg.]')

    # Done plotting everything, save the figure

     
    plt.tight_layout()
    if outfile is not None:
        plt.savefig(outfile, facecolor='white')
    else:
        plt.show()
        
    return plt

def plot_dof_timeseries_when_sampled(time, dihedrals, ene_traj, thresholds, outfile = None, title = None, idx_mins= None):
    """
    This function will plot the data related to all states, when those states are sampled.
    
    """
    nstates = len(dihedrals)

    active_qualitative_map_mligs = lambda num_ligs: plt.cm.viridis(np.linspace(0,1,num_ligs))
    
    colors = active_qualitative_map_mligs(nstates)
    
    
    # Split the different states in subplots:

    ncols = 4 if nstates > 11 else 3
    nrows = int(np.ceil(nstates/ncols))

    figsize = [3*ncols, 3*nrows]
    fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=figsize)
    
    if title is None:
        fig.suptitle('Dihedrals for optimized states', fontsize = 16)
    else:
        fig.suptitle(title, fontsize = 16)
    
    
    row_num = 0 # this starts at 0, and gets updated when col_num = ncols
    col_num = 0

    for i, dihedral in enumerate(dihedrals):
        if col_num == ncols-1: row_num +=1
        col_num = i % ncols
        state_num = i+1
        
        idx_sampled = np.where(ene_traj['e'+str(i+1)] < thresholds[i])
        
        axes[row_num, col_num].scatter(time[idx_sampled], mod2pi(dihedral[idx_sampled]), s = 4, color = colors[i], zorder = 2)
        
        
        axes[row_num, col_num].scatter(time, mod2pi(dihedral), s = 0.5, color = 'lightgrey', zorder = 1)
        
        axes[row_num, col_num].set_ylim([0, 360])
        axes[row_num, col_num].set_title('state ' + str(i+1))
        
        # Set the ticks to make sense:
        
        axes[row_num, col_num].set_yticks([0, 60, 120, 180, 240, 300, 360])
        
        #if idx_mins is not None:
        #    print (idx_mins[i])
        #    axes[row_num, col_num].scatter(time[idx_mins[i]], mod2pi(dihedral)[idx_mins[i]], 
        #                                    color = 'black', marker ='*', s = 32)
        
        
        
    # Add labels
    for i in range(ncols):
        axes[nrows-1, i].set_xlabel('time [ns]')
    
    for i in range(nrows):
        axes[i, 0].set_ylabel('dihedral angle [deg.]')

    # Done plotting everything, save the figure

     
    plt.tight_layout()
    if outfile is not None:
        plt.savefig(outfile, facecolor='white')
    else:
        plt.show()
        
    return plt

def find_idx_min(ene_traj, num_states = 13, num_seeds = 5):
    """
    We will find the minimum of each independent seed here:
    """
    
    # 1 :make a copy of the original data
    
    new_traj = copy.deepcopy(ene_traj)
    new_traj.reset_index(inplace=True)
    
    idx_mins = []
    
    s = len(ene_traj) / num_seeds
    for i in range(1, num_states+1):
        tmp =  []
        for j in range(0, num_seeds):            
            subtraj = new_traj[f'e{i}'][int(j*s):int((j+1)*s)]
            
            tmp.append(subtraj.idxmin())
            
        idx_mins.append(tmp)
        
    return idx_mins


def get_and_sort_trcs(root_path, num_seeds = 5):
    """
    This function returns a sorted list of all trc files 
    for which we can extract the data of interest to plot dihedrals for example
    
    """
    all_paths = []
    
    def get_s(path_trc):
        return int(path_trc.split('/')[-1].split('_')[-1].replace('.trc.gz', ''))
    
    for seed in range(1, num_seeds+1):
        paths = (root_path +f'/g_prod_seed{seed}/simulation/*/*.trc.gz')
    
        #print (paths)
    
        trcs = [p for p in glob.glob(paths) if 'eq_' not in p and get_s(p) == 1]
    
        for p in (sorted(trcs, key = lambda x : int(x.split('/')[-1].split('_')[-2]))):
            all_paths.append(p)
    
    return all_paths
```
