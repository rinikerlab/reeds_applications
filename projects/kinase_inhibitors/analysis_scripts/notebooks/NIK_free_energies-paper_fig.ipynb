# Load TI data

```python
def load_ti_data(path):
    
    dgs = np.zeros([2, 6])
    for i, leg in enumerate(['1_2', '1_3', '1_4', '1_5', '1_6']):
        tmp = np.zeros(5)
        for seed in [1, 2, 3, 4, 5]:
            with open(f'{path}/leg_{leg}/seed{seed}/ti_integral_averaged.txt', 'r') as f:
                last_line = f.readlines()[-1]
            
            tmp[seed-1] = float(last_line.split()[-1])
        
        dgs[0, i+1] = np.mean(tmp)
        dgs[1, i+1] = np.std(tmp)
    
    return dgs
    
```

```python
dg_ti_complex = load_ti_data(f'{prefix}/../ti_results/complex')
dg_ti_water = load_ti_data(f'{prefix}/../ti_results/solvent')

_, dgs_reti_openff_new = offset_comparison(ic50s, dg_ti_complex, dg_ti_water)
```

```python
def plot_errorbar(ax, x, y, y_err, color, label, zorder, alpha, marker = 's'):
    markers, caps, bars = ax.errorbar(x, y, y_err, fmt=marker, markersize = 6, 
                                      color = color, mec = color, 
                                      ecolor = color, label = label, 
                                      capthick=1.5, capsize=4, linewidth = 3, 
                                      zorder=zorder, alpha = 1
                                     )
    [bar.set_alpha(alpha) for bar in bars]
    [cap.set_alpha(alpha) for cap in caps]
        
    
from matplotlib import cm
active_qualitative_map_mligs = lambda num_ligs: plt.cm.viridis(np.linspace(0,1,num_ligs))
# active_qualitative_map_mligs = lambda num_ligs: plt.cm.cividis(np.linspace(0,1,num_ligs))
color_blind_frienfly = active_qualitative_map_mligs(5)


ax = axd['A']
fontsize = 18


exp = dgs_exp
sim = [dgs_sim_gaff, dgs_sim_openff]
labels_sim = ['GAFF', 'OpenFF']

plot_errorbar(ax, exp, dgs_sim_gaff[0], dgs_sim_gaff[1], 
              color = '#bb1177', label = labels_sim[0], 
              zorder = 12, alpha  = 1)



plot_errorbar(ax, exp, dgs_sim_openff[0], dgs_sim_openff[1], 
              color = '#3388cc',
              label = labels_sim[1], 
              zorder = 12, alpha  = 0.75
             )

# Test also adding TI to same plot? 

plot_errorbar(ax, exp, dgs_reti_openff_new[0], dgs_reti_openff_new[1], 
              color = 'darkorange',
              label = 'H-RE TI (OpenFF)',
              zorder = 11, alpha  = 0.75
             )


# Make a line 
x = np.linspace(-100, 100, 50)

ax.plot(x, x, color = 'black', zorder=5)
ax.plot(x, x-4.185, color = 'lightgrey', zorder=1)
ax.plot(x, x+4.185, color = 'lightgrey', zorder=1)

ax.fill_between(x, x-4.185, x+4.185, color='lightgrey', zorder=1)
ax.fill_between(x, x-2*4.185, x+2*4.185, color='lightgrey', alpha=0.5, zorder=1)

# Formatting
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1.5)

ax.tick_params(axis="y",direction="in", length = 6, width = 1.5, labelsize=fontsize)
ax.tick_params(axis="x",direction="in", length = 6, width = 1.5, labelsize=fontsize)

ax.set_xticks(np.arange(-65, -30,5))
ax.set_yticks(np.arange(-65, -30,5))

title = None

if title is not None:
    ax.set_title(title, fontsize = 18)

ax.set_xlabel(r'$\Delta G_{bind}$ - Experimental [kJ/mol]', fontsize = fontsize*0.90)
ax.set_ylabel(r'$\Delta G_{bind}$ - Simulation [kJ/mol]', fontsize = fontsize*0.90)

ax.set_ylim([-65, -32.5])
ax.set_xlim([-65, -32.5])
# ax.xaxis.set_tick_params(labelsize=20)

handles, labels = ax.get_legend_handles_labels()

print (handles)
print (labels)
order = [0,2,1]
leg = ax.legend([handles[idx] for idx in order],[labels[idx] for idx in order], 
               loc = 'lower center', fontsize=fontsize*.75, fancybox=True, ncol =2, 
                labelspacing=0.1, handletextpad = 0.2, borderpad = 0.3, 
               scatteryoffsets = [0.5])



#leg = ax.legend(loc = 'lower center', fontsize=fontsize*0.75, fancybox=True, ncol =2, 
#                labelspacing=0, handletextpad = 0.2, borderpad = 0.3)


leg.get_frame().set_edgecolor('black')

ax.set_facecolor('white')


for i in range(6):
    ax.annotate(f'N.{i+1}', (dgs_exp[i]+1, dgs_sim_openff[0][i]-1), weight='bold', fontsize = 12)
```
